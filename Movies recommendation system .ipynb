## **Now Setting up The `genres` Feature**

```python
data["genres"][0]
```

```python
def extracting_name(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
```

```python
data["cast"][0]
```

#### **Now Exctarcting only the First three actor name**

```python
def extracting_name_first_three_name(obj):
    counter=0
    l=[]
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i["name"])
            counter+=1
        else:
            break
    return l
```

```python
extracting_name_first_three_name(data["cast"][0])
```

```python
data["cast"]=data["cast"].apply(extracting_name_first_three_name)
```

```python
data["cast"][0]
```

### **Now Setting up the `crew`**

```python
data["crew"][0]
```

```python
def extracting_name_Dirctor(obj):
    counter=0
    l=[]
    for i in ast.literal_eval(obj):
        
        if i["job"]=="Director":
            l.append(i["name"])
            

    return l
```

```python
extracting_name_Dirctor(data["crew"][2])
```

```python
data["crew"]=data["crew"].apply(extracting_name_Dirctor)
```

# **Now Applay Vectorization On Tags columns**

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(stop_words="english",max_features=5000)
```

```python
cv
```

```python
vectors=cv.fit_transform(data["Tags"]).toarray()
```

```python
vectors
```

```python
vectors_DataFrame=pd.DataFrame(vectors)
vectors_DataFrame
```

#### **so for This we can appaly the steamming concepet on the Dataset**

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
data["Tags"].dtype
```

```python
def steming(text):
    Y=[]
    for i in text.split():
        Y.append(ps.stem(i))
    str=" ".join(Y)
    return str
```

```python
steming(data["Tags"][0])
```

```python
data["Tags"]=data["Tags"].apply(steming)
```

#### **Recommendation Funcation**

```python
def recommendation(Movies):
    Movies_index=data[data["title"]==Movies].index[0]
    Camparosin_movies=pd.DataFrame(similarity_btween_vectores_DataFrame.iloc[Movies_index])
    Camparosin_movies.sort_values(by=Movies_index,ascending=False,inplace=True)
    recommned_movies_index_list=Camparosin_movies.iloc[1:6].index
    Recomned_film=data.iloc[recommned_movies_index_list]
    Recomned_film=Recomned_film["title"].to_numpy()

    return Recomned_film
    

    
    

    
    
    
```
